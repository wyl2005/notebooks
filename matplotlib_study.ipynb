###调整X 或 Y轴上的刻度

```python
import random
plt.figure(figsize=(20,8))

x = range(120)
random.seed(10)
y = [random.uniform(20,35) for i in x]

plt.plot(x,y)

_x_ticks = ["10点{}分".format(i) for i in x if i< 60]
_x_ticks += ["11点{}分".format(i) for i in x if i < 60]

plt.xticks(x[::5], _x_ticks[::5], rotation=45)
plt.show()
```
