```python

data = fd["Rating"].values
max_data= data.max()
min_data= data.min()
#num_bin = (max_runtime - min_runtime) // 5

#设置不等宽组距, hist方法中取到一个左闭右开的区间[1.9, 3.5)
num_bin_list = [1.9, 3.5]
i = 3.5
while i<=max_data:
    i += 0.5
    num_bin_list.append(i)
    
print(num_bin_list)

plt.figure(figsize=(20,8),dpi=80)
plt.hist(data, num_bin_list)

#xticks让之前的组距能够对应上
plt.xticks(num_bin_list)
plt.show()
```
