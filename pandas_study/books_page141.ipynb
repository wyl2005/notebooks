```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./books.csv")

df.info()
print(df.head(1))
```

```python
data = df[df["original_publication_year"]]
```

```python
data = df[pd.notnull(df["original_publication_year"])]
data.info()
```

```python
grouped = data.groupby(by="original_publication_year")["average_rating"].mean()
#grouped = df.groupby(by="original_publication_year")
len(list(grouped))
print(type(grouped))
grouped
```

```python
_x = grouped.index
_y = grouped.values
_x
_x[::100]
```

```python
plt.figure(figsize=(20,8), dpi=80)
plt.plot(range(len(_x)), _y)

plt.xticks(list(range(len(_x)))[::10], _x[::10].astype(int), rotation=45)
plt.show()
```

```python
gp2 = data["average_rating"].groupby(by=data["original_publication_year"]).count()
#gp2 = data["average_rating"].groupby(by=data["original_publication_year"])
#list(gp2)[292]
```

```python
_x = gp2.index
_y = gp2.values
```

```python
plt.figure(figsize=(20,8), dpi=80)
plt.plot(range(len(_x)), _y)

plt.xticks(list(range(len(_x)))[::10], _x[::10].astype(int), rotation=45)
plt.show()
```
