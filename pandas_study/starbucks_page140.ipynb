```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./starbucks_store_worldwide.csv")
print(df.info())
#data1 = df.groupby(by="Country").count()["Brand"].sort_values(ascending=False)[:10]
data1 = df.groupby(by="Country").count()["Brand"].sort_values()[-10:]
#data1 = df.groupby(by="Country").count()[["Timezone","Brand"]].sort_values(by="Brand",ascending=False)[:10]

data1
```

```python
_x = data1.index
print(_x)
_y = data1.values
_y
```

```python
plt.figure(figsize=(20,8),dpi=80)

plt.bar(range(len(_x)), _y,width=0.4)
plt.xticks(range(len(_x)),_x)
plt.show()
```

```python
data2 = df[df["Country"]=="CN"].groupby(by="City")["Brand"].count().sort_values()[-50:]
print(data2)
_x2 = data2.index
_y2 = data2.values

plt.figure(figsize=(20,12),dpi=80)
#plt.bar(range(len(_x2)), _y2,width =0.3, color="orange")
#plt.xticks(range(len(_x2)), _x,rotation=45)

plt.barh(range(len(_x2)), _y2,height=0.3, color="orange")
plt.yticks(range(len(_x2)), _x,rotation=45)
plt.show()
```
